```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data=data[['Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','Credit_History','Property_Area','LoanAmount','Loan_Amount_Term']]
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Education']=encoder.fit_transform(data['Education'])
```

```python
data['Married']=encoder.fit_transform(data['Married'])
data['Self_Employed']=encoder.fit_transform(data['Self_Employed'])
data['Property_Area']=encoder.fit_transform(data['Property_Area'])
```

```python
data.head()
```

```python
data['Income']=data['ApplicantIncome']+data['CoapplicantIncome']
```

```python
data.drop(['CoapplicantIncome','ApplicantIncome'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Loan_Amount_Term'].unique()
```

```python
data.drop(['Loan_Amount_Term'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['Dependents'].unique()
```

```python
import re
data['Dependents']=data['Dependents'].apply(lambda s: re.sub(r'\+', '', s)).astype(int)
```

```python
X=data.drop(['LoanAmount'],axis=1)
```

```python
X.head()
```

```python
y=data[['LoanAmount']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=25)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("mse",mean_squared_error(y_test,y_pred))
print("mae",mean_absolute_error(y_test,y_pred))
print("r2",r2_score(y_test,y_pred))
```
